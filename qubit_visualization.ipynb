```python
import numpy as np
# Importing standard Qiskit libraries
from qiskit import QuantumCircuit, transpile, Aer, IBMQ, execute
from qiskit.tools.jupyter import *
from qiskit.visualization import plot_state_qsphere
from qiskit.visualization import plot_bloch_multivector
from ibm_quantum_widgets import *


# Loading your IBM Quantum account(s)
provider = IBMQ.load_account()
```

```python
#Create a simple circuit with just one qubit
qc = QuantumCircuit(1)
#Get state vector simulator backend
sv_simulator = Aer.get_backend('statevector_simulator')
#Run the circuit and get the state vector of the qubit
result = execute(qc, sv_simulator).result()
sv_results = result.get_statevector(qc)
```

```python
plot_bloch_multivector(sv_results)
```

```python
plot_state_qsphere(sv_results)
```

```python
#use not gate to flip the vector
qc.x(0)
#execute the circuit
result = execute(qc, sv_simulator).result()
sv_results = result.get_statevector(qc)
plot_state_qsphere(sv_results)
```

```python
#apply hadamarand gate
qc.h(0)
#execute the circuit
result = execute(qc, sv_simulator).result()
sv_results = result.get_statevector(qc)
plot_state_qsphere(sv_results)
```

```python
plot_bloch_multivector(sv_results)
```
